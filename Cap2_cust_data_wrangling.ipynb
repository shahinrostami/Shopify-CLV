Since we will be using the email address as a customer ID, let's drop them all to lowercase

```python
# changing email to lower case
df1['Email'] = df1['Email'].str.lower()
```

Before we anonymize the email addresses, let's extract the email server.

```python
## add email server as feature
df1['Server'] = df1['Email'].str.extract(r'\@(.*)')
```

```python
df1['Server'].value_counts().head(25)
```

We just noticed that in the top 25 email address, both the vendor and the holding company email servers show up. We should convert those accordingly, so they are hidden <br>


```python
df1['Server'].replace(to_replace='bambuearth.com', value='vendor', inplace=True)
df1['Server'].replace(to_replace='4x400.com', value='hold', inplace=True)
```

```python
# number of unique email domains
df1['Server'].unique().shape
```

All of the email addresses that are not in the top 25 would have less than 100 orders or less than 0.1% of the total orders. Each of these servers would be outliers in building a model feature. They are not useful for modeling. <br>
More importantly, some of these domain names could be used to identify small companies that have their own domain name email addresses. We are going to replace all of these with "custom" <br>

```python
# extract the most popular email domains and put them in a list
domain25 = df1['Server'].value_counts().head(25).index.tolist()
domain25
```

```python
# create a list of all unique domains
domains = df1['Server'].unique().tolist()
domains
```

```python
# remove the top 25 from the list of unique domains using list comprehension
dom = [x for x in domains if x not in domain25]
```

```python
# use the remaining list of unique domains to replace with value='custom'
df1['Server'].replace(to_replace=dom, value='custom', inplace=True)
```

Now, we need to remove some of the other columns that are identifiers

```python
# before dropping address information, let's create a boolean feature if Shipping Zip = Billing Zip
df1['ship_bill'] = df1['Shipping Zip'] == df1['Billing Zip']
```

```python
# let's look at shipping phone values
df1['Shipping Phone'].unique()
```

It looks like I can take the first 3 numbers to use as the area code using Regex if I wanted to extract the area code

```python
# let's look at the number of values for Shipping phones
df1['Shipping Phone'].unique().shape
```

```python
# let's extract the area code for the Shipping phone number to leave an area of indication
## add area code as a feature
df1['Area_Code'] = df1['Shipping Phone'].str.extract(r'^\(*([2-9]\d{2})')
```

```python
df1['Area_Code'].value_counts()
```

```python
# let's see if this filled all of the values
df1['Area_Code'].isnull().sum()
```

That's a lot of missing values. Let's use the billing phone number to fill some of those

```python
# extract billing area code first
df1['Bill_Area_Code'] = df1['Billing Phone'].str.extract(r'^\(*([2-9]\d{2})')
df1['Area_Code'].fillna(value=df1['Bill_Area_Code'], inplace=True)
df1['Area_Code'].isnull().sum()
```

That helped a little, but I think we can fill some more with the last "Phone" column

```python
# extract area code from general phone number
df1['Phone'] = str(df1['Phone'])
df1['Bill_Area_Code'] = df1['Phone'].str.extract(r'^\(*([2-9]\d{2})')
df1['Area_Code'].fillna(value=df1['Bill_Area_Code'], inplace=True)
df1['Area_Code'].isnull().sum()
```

Maybe that's as good as we can get

```python
# drop the shipping and billing information, except for the shipping zip code
df1.drop(['Billing Name', 'Billing Street', 'Billing Address1', 'Billing Address2', 'Billing Company',
       'Billing City', 'Billing Zip', 'Billing Province', 'Billing Country', 'Billing Phone', 'Shipping Name', 'Shipping Street',
       'Shipping Address1', 'Shipping Address2', 'Shipping Company', 'Shipping City', 'Shipping Province',
       'Shipping Country', 'Shipping Phone', 'Bill_Area_Code'], axis=1, inplace=True)
```

```python
# there may still be identifying features; potentially in vendor
df1['Vendor'].value_counts()
```

```python
# these need to be renamed to keep it anonymous
df1['Vendor'].replace(to_replace='Bambu Earth', value=0, inplace=True)
df1['Vendor'].replace(to_replace='Route', value=1, inplace=True)
```

```python
df1.info()
```

```python
# look at employee column to make sure that is not giving away any information
df1['Employee'].value_counts()
```

```python
# since they are actual employee names, we should convert them to booleans
df1['Employee'] = df1['Employee'].notnull()
# this should also replace all of the null values with False
```

```python
# phone number also looks like it could be something that could identify people
df1['Phone'].value_counts()
```

```python
# those are definitely real phone numbers. We already extracted area code from shipping phone (which was a more extensive list).
# we can drop this column
df1.drop(['Phone'], axis=1, inplace=True)
```

#### let's create customer ID's for each email address and then anonymize the email addresses

```python
df1['Email'].value_counts()
```

We have 39,679 unique emails <br>
Now, let's get this started. We'll search through the emails and make sure each email address has the same Cust_ID, so we can better track customers.

Let's actually look at using the existing customer ID's

```python
df1['Id'].unique().shape
```

```python
df1['Id'].value_counts()
```

Those look like they could be pretty good as identifiers, but to not mess with that column, we'll use the index

```python
# uses the first Id that appears by an email address as that unique identifier
df1['Cust_ID'] = df1.groupby('Email')['Id'].transform('first')
df1['Cust_ID'].value_counts()
```

The unique customer ID is equal to the number of email addresses. That's exactly what we wanted.

```python
# make sure that Cust_ID are the same when the email addresses are the same
df2 = df1
# we look for any cases where email addresses are the same but Cust_ID are different
df1[(df1['Email'] == df2['Email']) & (df1['Cust_ID'] != df2['Cust_ID'])]
```

That's extactly what I want to see - no duplicate customer numbers for the same email address <br>
